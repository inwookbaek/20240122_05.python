#### matplotlib 옵션

* http://matplotlib.org/examples/color/named_colors.html

```python
"""
========================
Visualizing named colors
========================

Simple plot example with the named colors and its visual representation.
"""
from __future__ import division

import matplotlib.pyplot as plt
from matplotlib import colors as mcolors


colors = dict(mcolors.BASE_COLORS, **mcolors.CSS4_COLORS)

# Sort colors by hue, saturation, value and name.
by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgba(color)[:3])), name)
                for name, color in colors.items())
sorted_names = [name for hsv, name in by_hsv]

n = len(sorted_names)
ncols = 4
nrows = n // ncols + 1

fig, ax = plt.subplots(figsize=(8, 5))

# Get height and width
X, Y = fig.get_dpi() * fig.get_size_inches()
h = Y / (nrows + 1)
w = X / ncols

for i, name in enumerate(sorted_names):
    col = i % ncols
    row = i // ncols
    y = Y - (row * h) - h

    xi_line = w * (col + 0.05)
    xf_line = w * (col + 0.25)
    xi_text = w * (col + 0.3)

    ax.text(xi_text, y, name, fontsize=(h * 0.8),
            horizontalalignment='left',
            verticalalignment='center')

    ax.hlines(y + h * 0.1, xi_line, xf_line,
              color=colors[name], linewidth=(h * 0.6))

ax.set_xlim(0, X)
ax.set_ylim(0, Y)
ax.set_axis_off()

fig.subplots_adjust(left=0, right=1,
                    top=1, bottom=0,
                    hspace=0, wspace=0)
plt.show()
```

##### Figure 객체
* 모든 그래프에는 Figure객체는 matplotlib.figure.Figure클래스 객체가 포함되어 있다.
* figure는 하나의 window를 가진다. Figure객체를 생성하려면 figure명령을 사용하지만 일반적을 plot()함수를 호출하면 Figure객체를 생성한다.
* figure명령을 사용하는 경우엔 여러개의 window를 동시에 띄우거나 figure크기를 설정할 할 경우에 사용한다.
* 한 개의 figure객체안에는 여러개의 그래프를 배열형태로 나타내는데 그 객체를 Axes객체라고 한다. 

##### Axes객체와 subplots()
* 여러개의 Axes객체를 생성하려면 subplot()를 호출해서 명시적으로 생성해야 한다.
* 하지만, plot명령을 사용하면 figure객체안에 axes객체를 생성해 준다.
* subplot()은 그리드(배열)형태로 axes객체를 생성한다.
* 주의할 점은 첫 번째의 axes의 순서는 0이 아니라 1부터 시작한다.

```python
# 1. subplot() : 2행 1열의 axes객체 생성하기
# plt.subplots? or plt.subplot?
# subplot(행,열,순서) or subplot(행열순서)
plt.figure(figsize=[10, 3])

x1 = np.linspace(0., 5.)
y1 = np.cos(2*np.pi*x1) * np.exp(-x1)
plt.subplot(211)
plt.plot(x1, y1, 'r-*')

x2 = np.linspace(0., 2.)
y2= np.cos(2*np.pi*x2)
plt.subplot(2,1,2)
plt.plot(x2, y2, 'g:o')

plt.tight_layout() # 그래프간의 간격을 자동으로 정렬해주는 함수
plt.show()
```

```python
# subplot(224) : 2행2열(1,2,3,4)
plt.figure(figsize=(10,3))

plt.subplot(221)
plt.title('1st axes - 221')
plt.plot(np.random.randn(10))

plt.subplot(222)
plt.title('2nd axes - 222')
plt.plot(np.random.randn(20))

plt.subplot(223)
plt.title('3rd axes - 223')
plt.plot(np.random.randn(25))

plt.subplot(224)
plt.title('4th axes - 224')
plt.plot(np.random.randn(10))

plt.tight_layout()
plt.show()
```

```python
# 1x3의 그래프그리기
# 1=bar, 2=sctter, 3=line

names = ['손흥민','김민재','이강인']
values = [50,60,70]

plt.figure(figsize=(10,3))

plt.subplot(131)
plt.title('1st axes - 131')
plt.bar(names, values)

plt.subplot(132)
plt.title('2nd axes - 132')
plt.scatter(names, values)

plt.subplot(133)
plt.title('3rd axes - 133')
plt.plot(names, values)

plt.suptitle('전체그래프의 제목')
plt.tight_layout()
plt.show()
```

#### 상자(box)그래프

* 기술통계학에서 box plot은 `수치적 통계자료(기술통계)를 표현`하는 그래프이다.
* box plot은 `5개의 요약수치를 표시`한다.
  1. 최소값    : 1사분위에서 1.5IQR값을 뺀 위치 
  2. 1사분위수 : 25%
  3. 2사분위수 : 50%, 중앙값(mean)
  4. 3사분위수 : 75%
  5. 최대값    : 3사분위에서 1.5IQR값을 더한 위치 
  6. 이상치    : 최소값, 최대값을 벗어난 값

```python
# 1. box plot
df = pd.DataFrame(np.random.rand(10,5), columns=['A','B','C','D','E'])
df.plot.box()
plt.show()
```

```python
# box plot 꾸미기
colors = dict(boxes='DarkGreen', whiskers='DarkOrange', medians='DarkBlue', caps='Black')

df.plot.box(color=colors)
plt.show()
```

```python
# 수평box plot
df.plot.box(color=colors, vert=False)
plt.show()

df.plot.box(color=colors, vert=False, positions=[10,4,5,6,8])
plt.show()
```

#### Area그래프

```python
# area()그래프
# 음수값이 있는 경우에는 에러 즉, area그래프는 음수값이 존재하지 않는다.
df = pd.DataFrame(np.random.rand(10,4), columns=['A','B','C','D'])
df.plot.area(figsize=(3,2))
plt.show()
```
