# 반복문(for)

```python
for 변수 in iterable:
    pass
```

````python
# 1. for문과 range()함수
for i in range(10):
    print(i, end=", ")
print()

for i in range(10, 20):
    print(i, end=", ")
print()

for i in range(10, 20, 3):
    print(i, end=", ")
````

````python
# 2. 일반 for 문
# 1~10까지의 합을 구하기
tot = 0
for i in range(1, 11):
    tot += i
print(f'1~10까지의 합 = {tot}')

# 3의 배수의 합
tot = 0
for i in range(1, 101, 3):
    tot += i
print(f'1~100까지 3의 배수의 합 = {tot}')
````

````python
# 3. for문과 시퀀스객체의 indexing/slicing
words = ['java','javascript','R','python','sql','windows','machinelearning','deeplearning', 'love']

# 실습. 
# 1) 각 단어의 길이를 출력하기
for word in words:
    print(f'{word}의 길이는 {len(word)}자 입니다!!')
print()

# 2) 5자리이상의 단어만 출력하기
for word in words:
    if(len(word)>=5):
        print(f'{word}의 길이는 {len(word)}자 입니다!!')
print()

# 3) 5자리이상의 단어로 구성된 list생성하기
l = []
for word in words:
    if(len(word)>=5):
        # l.append(word)  # 마지막에 추가
        l.insert(0, word)  # 맨 앞에 추가
print(l, len(l))
````

````python
# 4. len()와 range함수 동시 사용하기
for i in range(len(words)):
    print(i, words[i])

strs = '안녕하세요!! 반갑습니다!!'

# 실습. strs에서 !만 제외해서 출력하기
# len(), range()를 이용
for i in range(len(strs)):
    if strs[i] != '!':
        print(i, strs[i], end=", ")
print()

# range함수의 옵션값은 음수도 가능
print(list(range(-10, -1)))
print(list(range(-1, -100, -10)))
````

````python
# 5. 중첩for문
for i in range(2, 3):
    for j in range(2,4):
        print(f'중첩 for문 ({i}, {j})번째 실행!!')
````

````python
# 실습1. for문을 이용해서 구구단 만들기
# 임의 단을 입력받아서 해당 단의 구구단 출력하기
#  3 * 2 = 6, ..... 3 * 9 = 27
dan = int(input("단을 입력하세요 => "))
print(f'---- {dan}단 ----')
for i in range(2, 10):
    print(f'{dan} x {i} = {dan * i:2d}')
````

````python
# 실습2. 2단~9단까지 세로 출력하기
# 중첩for문을 이용
# -- 2단 -- 
# 2 * = 4
#  ....
# -- 9단 --
# 9 * 9 = 81
for dan in range(2, 10):
    print(f'---- {dan}단 ----')
    for j in range(2, 10):
        print(f'{dan} x {j} = {dan * j:2d}')  
````

````python
# 실습3. 2단~9단까지 세로 출력하기
# -- 2단 -- ....  -- 9단 --
# 2 x 2 = 4       9 x 2 = 18
#  ...
# 2 x 9 = 18      9 x 9 = 81
title = ""
for dan in range(2, 10):
    title = title + f' -- {dan}단 --'
print(title)

for j in range(2, 10):
    gugudan = ""
    for dan in range(2, 10):
        gugudan = gugudan + f' {dan} x {j} = {dan*j:2d}'
    print(gugudan)
````

##### list안에 for문 사용하기

1. list안에 한 개의 for문 - 1차원
```python
[표현식 for 변수 in iterable]
```

2. list안에 여러 개의 for문 - n차원 즉, 중첩for문
```python
[
    표현식 for 변수 in iterable if 조건
           for 변수 in iterable if 조건
           ...
]
```

````python
# 1. list안에 1개의 for문 사용하기

nums = [1,2,3,4,5]
print(nums * 3) # 반복연산자

# 1) [1,2,3,4,5] 3의 배수의 list를 생성
# a. 일반for문
result = []
for num in nums:
    result.append(num * 3)
print('1. 3의 배수 = ', result)

# b. list안에 for문 - for문안에 조건절이 없을 경우
result = [num * 3 for num in nums]
print('2. 3의 배수 = ', result)
print()

# 2) for문에 조건절이 있으 ㄹ경우
# 짝수만 출력
# a. 일반 for 문
result = []
for num in nums:
    if(num%2==0):
        result.append(num)
print('1. 짝수 = ', result) 

# b. list안의 for문 - 조건절이 있을 경우
# 짝수만 3의 배수로 처리
result = [num * 3 for num in nums if num%2==0]
print('2. 짝수의 3의 배수 = ', result) 
````

````python
# 2. list안에 중첩 for문 사용하기
# 구구단의 결과를 한 개의 list에 저장하기
# 1) list안에 조건이 없을 경우
result = [dan * i for dan in range(2, 10)
                  for i   in range(2, 10)
         ]
print(result)

# 2) for문에 조건이 있을 경우
# 짝수만 출력하기
result = [dan * i for dan in range(2, 10) if dan%2==0
                  for i   in range(2, 10) if i%2==0
         ]
print(result)
````
