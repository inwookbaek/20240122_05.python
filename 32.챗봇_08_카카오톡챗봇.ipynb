## 8.2 카카오톡 챗봇 연동

* 오픈빌더의 개념 : https://developers.kakao.com/docs/latest/ko/daum-search/dev-guide

### 8.2.1 시나리오와 블럭

* 오픈빌더에서는 `사용자의도(intent)를 처리하는 가장 작은 단위를 block`이라 한다.
* 이 `블럭들이 모여서 하나의 시나리오scenario로 구성`된다.
* 이렇게 카카오챗봇은 시나리오를 어떻게 구성하느냐에 따라 챗봇의 기능이 달라진다.
* 블럭에는 `사용자예상발화패턴이나 봇이 수행하는 액션과 응답형식을 정의`한다.
* 즉, 사용자발화가 입력되었을 때 블럭 내 발화패턴을 기반으로 사용자의도가 파악되어 최종적으로 1개의 블럭이 실행된다.
* 오픈빌더에서 생성된 모든 봇은 기본시나리오를 포함한다. 
* 사용자는 설계하는 봇에 따라 시나리오와 블럭을 추가할 수 있다.
* 본 강좌에서는 사용자발화처리를 직접제작한 쳇봇엔진에서 처리하기 때문에 시나리오를 추가하는 것은 다루지 않는다.
* 기본시나리오에는 3가지블럭이 제공된다.

|블럭|설명|
|:----:|:---------|
|웰컴블럭|사용자가 카카오톡태널 방에 최초로 입장했을 떄 발송하는 환영메시지를 정의 이 블럭은 `최초 1회만 작동`|
|폴백블럭|챗봇이 사용자발화의도를 이해하지 못했을 때 동작을 정의, 일반적으로 '이해를 잘 못했습니다'와 같은 메시지를 정의|
|탈출블럭|챗봇이 사용자에게 계속해서 되묻는 상황에서 탈출할 경우 사용자명령어를 정의|
||본 강의에서는 사용자발화를 챗봇엔진에서 처리하기 때문에 탈출블럭은 사용하지 않는다.|

### 8.2.2 스킬

* `스킬은 외부에서 구현한 기능을 호출해서 사용자에게 정보를 출력해 주는 기능`이다.
* 스킬 skill을 통해 챗봇 기술을 추가할 수 있다.
* 스킬은 블록의 출력기능과 유사하지만 `응답설정을 프로그램으로 처리할 수 있는 장점`이 있다.
* 즉, `외부에서 만든 기능을 카카오톡 챗봇에 맞게 응답을 출력`할 수 있다.
* 스킬도 해당 블럭의 발화에 반응하여 응답을 리턴하므로 `스킬과 블럭을 연결하는 과정이 필요`하다.
* 참고 : https://i.kakao.com/docs/key-concepts-skill#%EC%8A%A4%ED%82%AC%EC%9D%98-%EC%97%AD%ED%95%A0
* 스킬처리과정
  1. 사용자발화 : 11시에 자장면 주문할게요 
  1. 주문블럭   : 스킬호출 
  1. 스킬서버   : 주문정보를 DB에 저장하고 주문처리결과를 스킬응답 포맷에 맞게 리턴
  1. 챗봇       : 스킬서버의 응답을 사용자에게 리턴
  
* 본 강좌에서 만든 챗봇을 카카오톡 챗봇에 연결은?
  - 공식적으로는 오픈 빌더는 외부 챗봇엔진을 지원하지 않는다.
  - 하지만, 사용자발화의도를 이해하지 못했을 때 대응하는 `플백블록 full-back block을 지원`
  - `플백블럭은 사용자발화를 처리하는 시나리오블럭이 더 이상 없을 때 실행`된다.
  - 본 강좌는 모든 사요자 발화를 플백블럭에서 처리하도록 기본블럭 이외에 다른 시나리오 블럭은 생성하지 않는다.
  - 그다음 `플백블럭에서 챗봇API서버와 통신할 수 있는 스킬을 지정한다면 외부 챗봇엔진을 사용할 수 있다.`
  - 즉, `챗봇API서버를 스킬서버로 사용`한다.
  - 그림 1014 외부챗봇엔진 연동방법
  
## 8.3 스킬사용하기

* 카카오톡챗봇을 본강좌에서 제작한 `챗봇엔진에 연동하기 위해서는 스킬사용이 필수`이다.
* 스킬서버는 파이썬 Flask를 이용해서 제작

### 8.3.1 스킬서버 이해하기

* `스킬서버는 카카오봇시스템에서 받은 요청을 처리해 적절한 응답을 반환`한다.
* 각각의 요청은 HTTP POST메서드를 통해 전달되고 JSON형태로 데이터를 송수신한다.

### 8.3.3 스킬서버만들기

```python
!mkdir .\hello_kakao_skill
```

```python
%%writefile .\hello_kakao_skill\app.py
# 1. 스킬서버예제(1)

from flask import Flask, request
import sys

app = Flask(__name__)

@app.route('/api/sayHello', methods=['GET'])
def hello():
    return "Hello Kakao-Chatbot!!!"

# 1) 카카오톡 텍스트형태의 response
@app.route('/api/sayHello', methods=['POST'])
def sayHello():
    body = request.get_json()
    # print(body)

    responseBody = {
        "version": "2.0",
        "template": {
            "outputs": [
                {
                    "simpleText": {
                        "text": "안녕하세요? 카카오톡 챗봇입니다!!!"
                    }
                }
            ]
        }
    }

    return responseBody

# 2) 카카오톡 이미지형태의 response
@app.route('/api/showHello', methods=['POST'])
def showHello():
    body = request.get_json()
    
    responseBody = {
        "version": "2.0",
        "template": {
            "outputs": [
                {
                    "simpleImage": {
                        "imageUrl": "https://t1.daumcdn.net/friends/prod/category/M001_friends_ryan2.jpg",
                        "altText": "hello I'm Kakao.Chatbot!!"
                    }
                }
            ]
        }
    }

    return responseBody
    
if __name__ == '__main__':
    # app.run(host="0.0.0.0", port=5000) # web server
    app.run(host="0.0.0.0", port=int(sys.argv[1]), debug=True) # web server
```
